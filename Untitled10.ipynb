```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Convertir la columna 'fecha' al formato de fecha y hora
df['date'] = pd.to_datetime(df['date'])

# Filtrar para registros donde el nombre de la estación es 'Garfield-Dan Ryan'
Garfield_Dan_Ryan = df[df['stationname'] == 'Garfield-Dan Ryan']

# Establecer estilo marino y tamaño de figura.
sns.set(rc={'figure.figsize': (11, 4)})

# Trazar viajes en el tiempo para la estación Garfield-Dan Ryan
plt.plot(Garfield_Dan_Ryan['date'], Garfield_Dan_Ryan['rides'], linewidth=0.5)
plt.title('Viajes en el tiempo para la estación Garfield-Dan Ryan')
plt.xlabel('Fecha')
plt.ylabel('Número de viajes')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
df = pd.read_csv(url)

# Convertir la columna 'fecha' al formato de fecha y hora
df['date'] = pd.to_datetime(df['date'])

# Filtrar para registros donde el nombre de la estación es 'Garfield-South Elevated'
Garfield_South_Elevated = df[df['stationname'] == 'Garfield-South Elevated']

# Establecer estilo marino y tamaño de figura.
sns.set(rc={'figure.figsize': (11, 4)})

# Trazar viajes a lo largo del tiempo para la estación elevada Garfield-South
plt.plot(Garfield_South_Elevated['date'], Garfield_South_Elevated['rides'], linewidth=0.5)
plt.title('Viajes en el tiempo para la estación elevada Garfield-South')
plt.xlabel('fecha')
plt.ylabel('Number of Rides')
plt.grid(True)
plt.show()
```
